```python
#This is the solution of Q.No.4.

def find_factorial(number):
  if number < 0:
    return "Factorial is not defined for negative numbers"
  elif number == 0:
    return 1
  else:
    factorial = 1
    for i in range(1, number + 1):
      factorial *= i
    return factorial

num = 7
result = find_factorial(num)
print(f"The factorial of {num} is {result}")
```

```python
'''Q.No_14:- Write a program to find the maximum element in a 2D list using a nested `for` loop.''
```

```python
#This is the solution of Q.No.14.

def find_maximum_in_2d_list(data):
  if not data:
    return None

  maximum = data[0][0]  

  for row in data:
    for element in row:
      if element > maximum:
        maximum = element
  return maximum

# Example :
my_2d_list = [[12, 54, 34], [84, 27, 10], [74, 56, 59]]
max_element = find_maximum_in_2d_list(my_2d_list)
print(f"The maximum element is: {max_element}")

empty_2d_list = []
max_element_empty = find_maximum_in_2d_list(empty_2d_list)
print(f"The maximum element in the empty list is: {max_element_empty}")
```

```python
'''Q.No_20:- Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.''
```

```python
#This is the solution of Q.No.20.

def find_gcd(a, b):
    smaller = min(a, b)
    for i in range(1, smaller + 1):
        if (a % i == 0) and (b % i == 0):
            gcd = i  

    return gcd

num1 = 484
num2 = 258
gcd_result = find_gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is: {gcd_result}")
```

```python
'''Q.No_22:- Write a program to find the prime factors of a given number using a `for` loop and list comprehension.''
```

```python
#This is the solution of Q.No.22.

def prime_factors(n):
  factors = []
  while n % 2 == 0:
    factors.append(2)
    n //= 2
  for i in range(3, int(n**0.5) + 1, 2):
    while n % i == 0:
      factors.append(i)
      n //= i
  if n > 2:
    factors.append(n)

  return factors

number = 207
factors = prime_factors(number)
print(f"The prime factors of {number} are: {factors}")
```

```python
                                        #Challenge Level:-
'''Q.No_31:- Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list
comprehension.''
```

```python
#This is the solution of Q.No.31.

def primes_up_to(limit):
  is_prime = [True] * (limit + 1)
  is_prime[0] = is_prime[1] = False  

  for number in range(2, int(limit**0.5) + 1):
    if is_prime[number]:
      for multiple in range(number * number, limit + 1, number):
        is_prime[multiple] = False

  return [number for number, prime in enumerate(is_prime) if prime]

limit = 27
prime_list = primes_up_to(limit)
prime_list
```

```python
'''Q.No_32:- Create a program that generates a list of all Pythagorean triplets up to a specified limit using list
comprehension.''
```

```python
#This is the solution of Q.No.32.

def generate_pythagorean_triplets(limit):
  triplets = [(a, b, int((a**2 + b**2)**0.5))
              for a in range(1, limit + 1)
              for b in range(a, limit + 1)
              if (a**2 + b**2)**0.5 == int((a**2 + b**2)**0.5) and (a**2 + b**2)**0.5 <= limit]
  return triplets

limit = 38
pythagorean_triplets = generate_pythagorean_triplets(limit)
pythagorean_triplets
```

```python
'''Q.No_34:- Write a program that calculates the mean, median, and mode of a list of numbers using list
comprehension.''
```

```python
#This is the solution of Q.No.34.

from collections import Counter

def calculate_stats(numbers):
  if not numbers:
    return None, None, None
  mean = sum(numbers) / len(numbers)
  sorted_numbers = sorted(numbers)
  n = len(sorted_numbers)
  if n % 2 == 0:
    median = (sorted_numbers[n // 2 - 1] + sorted_numbers[n // 2]) / 2
  else:
    median = sorted_numbers[n // 2]
  counts = Counter(numbers)
  max_freq = max(counts.values())
  mode = [num for num, freq in counts.items() if freq == max_freq]
  if len(mode) == len(numbers):
      mode = "No unique mode"
  elif len(mode) == 1:
      mode = mode[0]


  return mean, median, mode

data = [11,42,73, 47, 57, 85, 56, 87, 18, 58, 48,]
mean_val, median_val, mode_val = calculate_stats(data)

print(f"Data: {data}")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

```python
#This is the solution of Q.No.35.

def pascals_triangle(rows):
    if rows <= 0:
        return []

    triangle = [[1]]
    for i in range(1, rows):
        row = [1] + [triangle[i-1][j-1] + triangle[i-1][j] for j in range(1, i)] + [1]
        triangle.append(row)
    return triangle

num_rows = 5
pascal_triangle_result = pascals_triangle(num_rows)
for row in pascal_triangle_result:
    row
```

```python
#This is the solution of Q.No.40.

import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def is_palindrome(n):
    return str(n) == str(n)[::-1]

prime_palindromes = [n for n in range(2, 505) if is_prime(n) and is_palindrome(n)]

prime_palindromes
```
