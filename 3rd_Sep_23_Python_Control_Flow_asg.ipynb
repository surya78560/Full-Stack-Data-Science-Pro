```python
#This is the solution of Q.No.12.

def classify_triangle(side1, side2, side3):
    if not (side1 + side2 > side3 and side1 + side3 > side2 and side2 + side3 > side1):
        return "Invalid Triangle"

    if side1 == side2 == side3:
        return "Equilateral"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles"
    else:
        return "Scalene"

side_a = 7
side_b = 8
side_c = 7
print(f"Triangle with sides {side_a}, {side_b}, {side_c} is: {classify_triangle(side_a, side_b, side_c)}")
```

```python
#This is the solution of Q.No.13.

def check_year(year):
  is_century_year = year % 100 == 0
  is_leap_year = (year % 4 == 0 and not is_century_year) or (is_century_year and year % 400 == 0)
  return is_leap_year, is_century_year

year = 2055
is_leap, is_century = check_year(year)

if is_leap and is_century:
  print(f"{year} is a leap year and a century year.")
elif is_leap:
  print(f"{year} is a leap year but not a century year.")
elif is_century:
  print(f"{year} is a century year but not a leap year.")
else:
  print(f"{year} is neither a leap year nor a century year.")

year = 1900
is_leap, is_century = check_year(year)

if is_leap and is_century:
  print(f"{year} is a leap year and a century year.")
elif is_leap:
  print(f"{year} is a leap year but not a century year.")
elif is_century:
  print(f"{year} is a century year but not a leap year.")
else:
  print(f"{year} is neither a leap year nor a century year.")
```

```python
'''Q.No_17:- Write a Python program to calculate the roots of a quadratic equation.''
```

```python
#This is the solution of Q.No.17.

import cmath

def find_roots_quadratic(a, b, c):
    delta = (b**2) - 4*(a*c)
    root1 = (-b - cmath.sqrt(delta)) / (2 * a)
    root2 = (-b + cmath.sqrt(delta)) / (2 * a)

    return (root1, root2)

# For x^2 - 5x + 6 = 0
a = 5
b = -6
c = 9
roots = find_roots_quadratic(a, b, c)
print(f"The roots of the equation {a}x^2 + {b}x + {c} = 0 are: {roots}")
```

```python
'''Q.No_20:- Develop a program that checks if a given number is prime or not using nested if-else statements.''
```

```python
#This is the solution of Q.No.20.

def is_prime_nested(number):
  if number <= 1:
    return False  # Numbers less than or equal to 1 are not prime

  if number > 1:
    for i in range(2, int(number**0.5) + 1):
      if (number % i) == 0:
        return False  
    return True

num = int(input("Enter an integer: "))

if is_prime_nested(num):
  print(f"{num} is a prime number.")
else:
  print(f"{num} is not a prime number.")
```

```python
'''Q.No_26:- Implement a program to calculate the discounted price based on different purchase amounts using elif
statements.''
```

```python
#This is the solution of Q.No.26.

def calculate_discounted_price(purchase_amount):
  if purchase_amount >= 1000:
    discount_rate = 0.20
  elif purchase_amount >= 500:
    discount_rate = 0.10 
  elif purchase_amount >= 100:
    discount_rate = 0.05 
  else:
    discount_rate = 0.0 

  discounted_price = purchase_amount * (1 - discount_rate)
  return discounted_price

# Example usage:
purchase1 = 1575
discounted1 = calculate_discounted_price(purchase1)
print(f"Purchase amount: ${purchase1}, Discounted price: ${discounted1:.2f}")
```

```python
'''Q.No_27:- Develop a program to calculate the electricity bill based on different consumption slabs using elif
statements.''
```

```python
#This is the solution of Q.No.27.

def calculate_electricity_bill(consumption):
  bill = 0
  if consumption <= 50:
    bill = consumption * 3.50
  elif consumption <= 150:
    bill = 50 * 3.50 + (consumption - 50) * 4.00
  elif consumption <= 250:
    bill = 50 * 3.50 + 100 * 4.00 + (consumption - 150) * 5.20
  else:
    bill = 50 * 3.50 + 100 * 4.00 + 100 * 5.20 + (consumption - 250) * 6.50

  return bill

# Example usage:
units_consumed = 254
total_bill = calculate_electricity_bill(units_consumed)
print(f"The electricity bill for {units_consumed} units is: Rs. {total_bill:.2f}")
```

```python
'''Q.No_28:- Create a program to determine the type of quadrilateral based on its angles and sides using elif
statements.''
```

```python
#This is the solution of Q.No.28.

def determine_quadrilateral(angle1, angle2, angle3, angle4, side1, side2, side3, side4):
  angles_sum = angle1 + angle2 + angle3 + angle4
  if angles_sum != 360:
    return "Not a quadrilateral (angles don't sum to 360)"

  if side1 == side2 == side3 == side4:
    if angle1 == angle2 == angle3 == angle4 == 90:
      return "Square"
    else:
      return "Rhombus"
  elif side1 == side3 and side2 == side4:
    if angle1 == angle2 == angle3 == angle4 == 90:
      return "Rectangle"
    else:
      return "Parallelogram"
  elif (side1 == side2 and side3 == side4) or (side1 == side4 and side2 == side3):
    return "Kite"
    return "Quadrilateral"

# Example usage:
angle1 = 90
angle2 = 80
angle3 = 75
angle4 = 98
side1 = 7
side2 = 5
side3 = 8
side4 = 4

quadrilateral_type = determine_quadrilateral(angle1, angle2, angle3, angle4, side1, side2, side3, side4)
print(f"The quadrilateral is a: {quadrilateral_type}")
```
