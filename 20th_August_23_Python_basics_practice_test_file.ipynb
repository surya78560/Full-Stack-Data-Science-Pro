```python
#This is the solution of Q.NO.42

def caesar_cipher(text,shift):
    
  """"
    Encodes a string using a Caesar cipher.

    Args:
      text: The string to encode.
      shift: The number of positions to shift
    Returns:
      The encoded string.
  """
  encoded_text = ""
  # The for loop was indented too far. It should be aligned with the previous line.
  for char in text:
    if char.isalpha():
      encoded_char = chr((ord(char) + shift - ord('a')) % 26 + ord('a'))
    else:
      encoded_char = char
    encoded_text += encoded_char
  return encoded_text


text = "Surya,Singh!"
shift = 3
print(f"{caesar_cipher(text,shift)}")
```

```python
''Q.NO.43:- Decode a Caesar cipher encoded string.''
```

```python
#This is the solution of Q.NO.43

def caesar_cipher_decoder(text,shift):
    """
    Decodes a string using a Caesar cipher.

    Args:
        text: The string to decode.
        shift: The number of positions to shift each letter.

    Returns:
        The decode string.
    """
    decoded_text = ""
    for char in text:
        if char.isalpha():
            decoded_char = chr((ord(char) - shift - ord('a')) % 26 + ord('a'))
        else:
            decoded_char = char
    decoded_text += decoded_char
    return decoded_text

text = "SINGH, SURYAA!"
shift = 3
print(f"{caesar_cipher_decoder(text,shift)}")
```

```python
''Q.NO.44:- Find the most frequent word in a string.''
```

```python
#This is the solution of Q.NO.44

import re
from collections import Counter

def most_frequent_word(input_string):
    clean_string = re.sub(r'[^\w\s]','',input_string).lower()
    words = clean_string.split()
    word_counter = Counter(words)
    most_common_word = word_counter.most_common(1)[0][0]
    return most_common_word

input_string = "Hello Surya, are you there! How's you?"
most_common = most_frequent_word(input_string)
print(f"Most Frequent word:- '{most_common}'")
```

```python
''Q.NO.34:- Split a list into chunks of size `n`.''
```

```python
#This is the solution of Q.NO.34.

def sieve_of_eratosthenes(limit):
    primes = []
    is_prime = [True] * (limit + 1)
    is_prime[0] = is_prime[1] = False

    for num in range(2, int(limit ** 0.5) + 1):
        if is_prime[num]:
            primes.append(num)
            for multiple in range(num * num, limit + 1, num):
                is_prime[multiple] = False

    for num in range(int(limit ** 0.5) + 1, limit + 1):
        if is_prime[num]:
            primes.append(num)

    return primes

prime_numbers = sieve_of_eratosthenes(70)

print("List of prime numbers up to 70:-", prime_numbers)
```

```python
#This is the solution of Q.NO.39.

def rotate_tuple_left(tup, n):
    n = n % len(tup)  # Ensure n is within the range of tuple length
    rotated_tuple = tup[n:] + tup[:n]  # Rotate the tuple
    return rotated_tuple

my_tuple = (14, 23, 32, 42, 55)
n_positions = 3
result = rotate_tuple_left(my_tuple, n_positions)
print(result)
```

```python
''Q.NO.40:- Rotate a tuple to the right by `n` positions.''
```

```python
#This is the solution of Q.NO.40.

def rotate_tuple_right(tup, n):
    n = n % len(tup)  # Ensure n is within the range of tuple length
    rotated_tuple = tup[-n:] + tup[:-n]  # Rotate the tuple
    return rotated_tuple

my_tuple = (12, 22, 43, 34, 55,50)
n_positions = 3
result = rotate_tuple_right(my_tuple, n_positions)
print(result)
```

```python
#This is the solution of Q.NO.49.

def contains_subsequence(main_tuple, sub_tuple):
    sub_length = len(sub_tuple)
    for i in range(len(main_tuple) - sub_length + 1):
        if main_tuple[i:i+sub_length] == sub_tuple:
            return True
    return False

main_tuple = ( 6, 10, 15, 21, 28, 36)

sub_tuple = (1, 3, 6)

contains_subseq = contains_subsequence(main_tuple, sub_tuple)

print("Main tuple contains subsequence tuple:", contains_subseq)
```
